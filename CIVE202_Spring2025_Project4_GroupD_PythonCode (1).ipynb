```python
#Import Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

nri_dictonary = pd.read_csv('NRIDataDictionary.csv')
ne_nri = pd.read_csv('Nebraska Data.csv')
ne_svi = pd.read_csv('NebraskaSVI.csv')
io_nri = pd.read_csv('Iowa Data.csv')
io_svi = pd.read_csv('IowaSVI.csv')

print(io_svi)
print(ne_svi)
```

```python
#Cleaning Iowa NRI Data

print(io_nri.isna())

data = {'col1': [1, 2, np.nan, 4], 'col2': [5, np.nan, 7, 8]}
df = pd.DataFrame(data)
print(df)
print(df.isna())

io_fill_med = io_nri.fillna(0)
io_dropped_cols=io_fill_med.dropna()
io_nri_clean = io_dropped_cols
io_dropped_cols.to_csv('io_nri_med.csv', index = False)
```

```python
#Cleaning Nebraska NRI Data

print(ne_nri.isna())

data = {'col1': [1, 2, np.nan, 4], 'col2': [5, np.nan, 7, 8]}
df = pd.DataFrame(data)
print(df)
print(df.isna())

ne_fill_med = ne_nri.fillna(0)
ne_dropped_cols=ne_fill_med.dropna()
ne_nri_clean = ne_dropped_cols
ne_nri_clean.to_csv('ne_nri_med.csv', index = False)
```

```python
#Cleaning Iowa SVI Data

print(io_svi.isna())

data = {'col1': [1, 2, np.nan, 4], 'col2': [5, np.nan, 7, 8]}
df = pd.DataFrame(data)
print(df)
print(df.isna())

io_fill_med = io_svi.fillna(0)
io_dropped_cols=io_fill_med.dropna()
io_svi_clean = io_dropped_cols
io_svi_clean.to_csv('io_svi_med.csv', index = False)
```

```python
# Cleaning Nebraska SVI data

print(ne_svi.isna())

data = {'col1': [1, 2, np.nan, 4], 'col2': [5, np.nan, 7, 8]}
df = pd.DataFrame(data)
print(df)
print(df.isna())

ne_fill_med = ne_svi.fillna(0)
ne_dropped_cols=ne_fill_med.dropna()
ne_svi_clean = ne_dropped_cols
ne_svi_clean.to_csv('ne_svi_med.csv', index = False)
```

```python
# Renaming columns to get them to combine easy

io_nri_clean['STCNTY'] = io_nri_clean['STCOFIPS']
io_nri_clean.columns

ne_nri_clean['STCNTY'] = ne_nri_clean['STCOFIPS']
ne_nri_clean.columns
```

```python
# Merging Data

complete_data_ne = pd.merge(ne_nri_clean, ne_svi_clean, on = 'STCNTY')
complete_data_io = pd.merge(io_nri_clean, io_svi_clean, on = 'STCNTY')
print(complete_data_ne)
```

```python
# Merge without repeat names

complete_data_ne = pd.merge(ne_nri_clean, ne_svi_clean,on = 'STCNTY',suffixes = ('', '_right'))
complete_data_ne.columns
complete_data_io = pd.merge(io_nri_clean, io_svi_clean, on = 'STCNTY',suffixes = ('','_right'))
complete_data_io.columns

complete_data_ne.to_csv('Nebraska Data Merged.csv', index = False)
complete_data_io.to_csv('Iowa Data Merged.csv', index = False)
```

```python
# Creating new columns
# Variable: Lighting
# Frequency Variable: lighting frequency
# Impact Variable: lightning exposure - impacted area (sq mile)
complete_data_ne['Lightning Risk'] = complete_data_ne['LTNG_AFREQ']*complete_data_ne['LTNG_EXP_AREA']
print(complete_data_ne['Lightning Risk'])

# Variable: Tornado
#Frequency Variable: Tornado frequency
# Impact Variable: Expected Annual Loss- Agricultural Value
complete_data_ne['Tornado'] = complete_data_ne['TRND_AFREQ']*complete_data_ne['TRND_EALA']
print(complete_data_ne['Tornado'])

# Iowa with the same variables
complete_data_io['Lightning Risk'] = complete_data_io['LTNG_AFREQ']*complete_data_io['LTNG_EXP_AREA']
print(complete_data_io['Lightning Risk'])

complete_data_io['Tornado'] = complete_data_io['TRND_AFREQ']*complete_data_io['TRND_EALA']
print(complete_data_io['Tornado'])
```

```python
# Group Data by Nebraska County

grouped_data_ne = complete_data_ne.groupby('COUNTY',as_index = False)

grouped_data_ne = complete_data_ne.groupby('COUNTY').agg(mean_lightning=('Lightning Risk', 'mean'),
                                                      mean_tornado=('Tornado','mean'))
print(grouped_data_ne)
                         
```

```python
# Create Bar chart

grouped_data_ne.plot(kind='bar', figsize=(20, 10), width=0.8)

# Customizing the plot
plt.xlabel('County')
plt.ylabel('Mean Risk Value')
plt.title('Comparison of Risk of Lightning vs Tornado')
plt.xticks(rotation=90)

# Show the plot
plt.legend(title="Risk Type")  # Adding a legend to distinguish risk types
plt.show()
```

```python
# Group Iowa Data by County

grouped_data_io = complete_data_io.groupby('COUNTY',as_index = False)

grouped_data_io = complete_data_io.groupby('COUNTY').agg(mean_lightning=('Lightning Risk', 'mean'),
                                                      mean_tornado=('Tornado','mean'))
print(grouped_data_io)
```

```python
# Create Bar Chart

# Create bar plot with side-by-side bars (dodged effect)
grouped_data_io.plot(kind='bar', figsize=(20, 10), width=0.8)

# Customizing the plot
plt.xlabel('County')
plt.ylabel('Mean Risk Value')
plt.title('Comparison of Risk of Lightning vs Tornado')
plt.xticks(rotation=90)

# Show the plot
plt.legend(title="Risk Type")  # Adding a legend to distinguish risk types
plt.show()
```
